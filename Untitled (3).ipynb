```python
import torch
import torch.nn as nn
import torch.optim as optim

# --------------------------
# 1. Example dataset
# --------------------------
# Suppose we have 5572 samples and 1000 features
X = torch.randn(5572, 1000)  # input features
y = torch.randn(5572, 1)     # targets for regression

# --------------------------
# 2. Define model
# --------------------------
class SimpleModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(SimpleModel, self).__init__()
        self.fc1 = nn.Linear(input_dim, 128)   # hidden layer
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(128, output_dim)  # output layer

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x

model = SimpleModel(input_dim=1000, output_dim=1)  # input matches X features

# --------------------------
# 3. Loss and optimizer
# --------------------------
loss_fn = nn.MSELoss()                      # regression loss
optimizer = optim.Adam(model.parameters(), lr=0.001)

# --------------------------
# 4. Training loop
# --------------------------
num_epochs = 1000
for i in range(num_epochs):
    optimizer.zero_grad()
    outputs = model(X)
    loss = loss_fn(outputs, y)
    loss.backward()
    optimizer.step()

    if i % 100 == 0:
        print(f"Epoch {i}: Loss = {loss.item():.4f}")

# --------------------------
# 5. Evaluation
# --------------------------
model.eval()  # switch to evaluation mode
with torch.no_grad():
    y_pred = model(X)
    print("Sample predictions:\n", y_pred[:5])
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# ----------------------------
# 1. Dummy dataset
# ----------------------------
# 100 samples, 10 features
X = torch.randn(100, 10)

# Binary labels (0 or 1)
y = (torch.rand(100) > 0.5).int()

# ----------------------------
# 2. Define a simple model
# ----------------------------
class SimpleNN(nn.Module):
    def __init__(self, input_dim):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_dim, 16)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(16, 1)

    def forward(self, x):
        return self.fc2(self.relu(self.fc1(x)))

model = SimpleNN(input_dim=10)

# ----------------------------
# 3. Train briefly (just to give non-random weights)
# ----------------------------
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

for epoch in range(20):
    optimizer.zero_grad()
    outputs = model(X).squeeze()
    loss = criterion(outputs, y.float())
    loss.backward()
    optimizer.step()

# ----------------------------
# 4. Evaluation
# ----------------------------
model.eval()
with torch.no_grad():
    y_pred = (torch.sigmoid(model(X).squeeze()) > 0.5).int()

    # Accuracy
    accuracy = (y_pred == y).float().mean()

    # Sensitivity (Recall for positives only)
    sensitivity = (y_pred[y == 1] == y[y == 1]).float().mean()

print("Accuracy:", accuracy.item())
print("Sensitivity:", sensitivity.item())
```

```python
import torch
import torch.nn as nn
from sklearn.feature_extraction.text import CountVectorizer

# Example messages
custom_messages = [
    "winner! Great deal, call us to get this product for free", 
    "Tomorrow is my birthday, do you come to the party?"
]

# Convert text → numeric vectors
vectorizer = CountVectorizer()
X_custom = vectorizer.fit_transform(custom_messages)  # sparse matrix
X_custom = torch.tensor(X_custom.toarray(), dtype=torch.float32)

# Dummy model (input_dim = vocab size)
class SimpleNN(nn.Module):
    def __init__(self, input_dim):
        super(SimpleNN, self).__init__()
        self.fc = nn.Linear(input_dim, 1)
    def forward(self, x):
        return self.fc(x)

model = SimpleNN(input_dim=X_custom.shape[1])

# Evaluate predictions
model.eval()
with torch.no_grad():
    pred = torch.sigmoid(model(X_custom))
    print(pred)
```
